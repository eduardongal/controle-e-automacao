**Simulação 1: Controle de Direção de Veículo com Esteiras**

**Objetivo:** Analisar a resposta ao degrau do sistema de controle de direção de um veículo com esteiras para verificar a estabilidade e o desempenho do sistema.

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Parâmetros dados
K = 70
a = 0.6

# Função de transferência para o sistema de controle de direção de veículo com esteiras
num1 = [K, K*a]
den1 = [1, 8, 17, K+10, K*a]
system1 = ctrl.TransferFunction(num1, den1)

# Vetor de tempo para a simulação
t = np.linspace(0, 10, 1000)

# Resposta ao degrau para o sistema de controle de direção de veículo com esteiras
t1, y1 = ctrl.step_response(system1, T=t)

# Plotando a resposta ao degrau
plt.figure()
plt.plot(t1, y1)
plt.title('Resposta ao Degrau do Sistema de Controle de Direção de Veículo com Esteiras')
plt.xlabel('Tempo (s)')
plt.ylabel('Resposta')
plt.grid(True)
plt.show()
```

**Simulação 2: Motocicleta Controlada por Robô**

**Objetivo:** Analisar a resposta ao degrau do sistema de motocicleta controlada por robô para verificar a estabilidade e o desempenho do sistema em manter a motocicleta na posição vertical.

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Parâmetros dados
a1 = 9
a2 = 2.7
a3 = 1.35
t = 0.2
Kp = 10
Kd = 5

# Função de transferência para o sistema de motocicleta controlada por robô
num2 = [a2, a3]
den2 = [t, (1 + Kd * a3), (Kd * a2 + Kp * a3 - t * a1), (Kp * a2 - a1)]
system2 = ctrl.TransferFunction(num2, den2)

# Resposta ao degrau para o sistema de motocicleta controlada por robô
t2, y2 = ctrl.step_response(system2, T=t)

# Plotando a resposta ao degrau
plt.figure()
plt.plot(t2, y2)
plt.title('Resposta ao Degrau do Sistema de Motocicleta Controlada por Robô')
plt.xlabel('Tempo (s)')
plt.ylabel('Resposta')
plt.grid(True)
plt.show()
```
